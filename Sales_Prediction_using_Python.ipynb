```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create a bar graph for TV, radio, and newspaper advertising spending
plt.figure(figsize=(10, 6))
plt.bar(['TV', 'Radio', 'Newspaper'], [Advertising['TV'].sum(), Advertising['Radio'].sum(), Advertising['Newspaper'].sum()])
plt.xlabel('Advertising Medium')
plt.ylabel('Total Spending')
plt.title('Total Advertising Spending per Medium')
plt.show()

# Create a scatter plot to show the relationship between sales and advertising spending for each medium
plt.figure(figsize=(12, 6))
plt.scatter(Advertising['TV'], Advertising['Sales'], label='TV')
plt.scatter(Advertising['Radio'], Advertising['Sales'], label='Radio')
plt.scatter(Advertising['Newspaper'], Advertising['Sales'], label='Newspaper')
plt.xlabel('Advertising Spending')
plt.ylabel('Sales')
plt.title('Sales vs. Advertising Spending')
plt.legend()
plt.show()
```
